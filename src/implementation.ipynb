## Macros Scheme

Divisão entre as primitivas:

- Cena
- Quadro
- Cenario (Background)
- Personagem
- Efeito

```python
;; =========================
;; HELPERS
;; =========================

(define (normalize str)
  (let* ((chars (string->list str))
         (filtered
          (map (lambda (c)
                 (cond
                   ((char-alphabetic? c) (char-downcase c))
                   ((char-numeric? c) c)
                   ((char-whitespace? c) #\_)
                   (else #\_)))
               chars)))
    (list->string filtered)))

(define (resolve-background name)
  (string-append "assets/backgrounds/" (normalize name) ".png"))

(define (resolve-character name emotion)
  (string-append
    "assets/characters/"
    (normalize name)
    "/"
    (normalize emotion)
    ".png"))

(define (resolve-effect name)
  (string-append "assets/effects/" (normalize name) ".png"))

;; =========================
;; MACROS
;; =========================

(define-syntax CENA
  (syntax-rules ()
    ((_ nome (Local local) quadros ...)
     (string-append
      "<div class='scene'>"
      "<h2>" nome " - " local "</h2>"
      quadros ...
      "</div>")))
)

(define-syntax QUADRO
  (syntax-rules ()
    ((_ numero (Layout layout) elementos ...)
     (string-append
      "<div class='panel'>"
      elementos ...
      "</div>")))
)

(define-syntax CENARIO
  (syntax-rules ()
    ((_ nome)
     (string-append
      "<div class='background' style='background-image:url(\""
      (resolve-background nome)
      "\")'></div>")))
)

(define-syntax RECORDATORIO
  (syntax-rules ()
    ((_ texto)
     (string-append
      "<div class='caption'>" texto "</div>")))
)

(define-syntax EFEITO
  (syntax-rules ()
    ((_ nome)
     (string-append
      "<div class='effect'>"
      "<img src='"
      (resolve-effect nome)
      "'/>"
      "</div>")))
)

(define-syntax PERSONAGEM
  (syntax-rules ()
    ((_ nome (Posicao pos) (Sentimento emo))
     (string-append
      "<div class='character'>"
      "<img src='"
      (resolve-character (symbol->string 'nome) emo)
      "'/>"
      "<p>" (symbol->string 'nome) " (" emo ")</p>"
      "</div>")))
)

(define-syntax FALA
  (syntax-rules ()
    ((_ personagem acao texto)
     (string-append
      "<div class='speech'>"
      "<b>" (symbol->string 'personagem)
      " (" (symbol->string 'acao) ")</b>: "
      texto
      "</div>")))
)

;; =========================
;; HTML BASE
;; =========================

(define (wrap-html content)
  (string-append
   "<html><head><style>"
   "body { font-family: Arial; background:#111; color:#fff; }"
   ".panel { width:500px; height:500px; position:relative; overflow:hidden; border:2px solid #fff; margin:10px; }"
   ".background { position:absolute; inset:0; background-size:cover; }"
   ".character { position:absolute; bottom:20px; left:20px; z-index:1; }"
   ".speech { position:absolute; top:20px; right:20px; background:white; color:black; padding:10px; border-radius:10px; z-index:2; }"
   ".effect { position:absolute; top:50px; right:50px; z-index:2; }"
   ".caption { background:black; padding:5px; }"
   "</style></head><body>"
   content
   "</body></html>"))

;; =========================
;; USO (SUA DSL EM SCHEME)
;; =========================

(define resultado
  (wrap-html
   (CENA "O Despertar do Poder"
     (Local "Floresta")
     (QUADRO 1
       (Layout "Topo_Largo")
       (CENARIO "Floresta")
       (PERSONAGEM Joao (Posicao "Esquerda") (Sentimento "Em guarda"))
       (FALA Joao pensa "Eu sinto uma presença... ela está por perto.")
       (EFEITO "CRACK!")
       (RECORDATORIO "O silêncio da floresta é interrompido.")
     )
   )
  )
)

(display resultado)
```
